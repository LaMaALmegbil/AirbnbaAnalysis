**Cleaning the dataset**


Some columns have been deleted because they do not help us answer our questions, such as: host_id,host_name,neighbourhood_group,last_review,reviews_per_month,latitude and longitude.

```python
df.duplicated().sum()
```

```python
df.drop(columns =['host_id', 'host_name','neighbourhood_group','last_review','reviews_per_month','latitude','longitude'],inplace=True)
```

```python
df
```

"has_license" column has been added which take boolean value true or false.

```python
df['has_license'] = df['license'].notna().map({True: 'yes', False: 'no'})
```

```python
df
```

The null values in the price are filled with the mean of the price column.

```python
mean=df['price'].mean()
df['price']=df['price'].fillna(mean)
```

```python
df.info()
```
